```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('titanic.csv')
```

```python
df.sample(5)
```

```python
df['number'].unique()
```

```python
figure = df['number'].value_counts().plot.bar()
figure.set_title('Passengers Travelling With')
```

```python
# extract numeric part
df['number_numerical'] = pd.to_numeric(df["number"],errors = 'coerce', downcast='integer')
```

```python
# extract categorical part
df['number_categorical'] = np.where(df['number_numerical'].isnull(),df['number'],np.nan)

df.head()
```

```python
df['Cabin'].unique()
```

```python
df['cabin_num'] = df['Cabin'].str.extract('(\d+)')  # to capture numeric part
df['cabin_cat'] = df['Cabin'].str[0]    # to capture categorical part

df.head()
```

```python
df['cabin_cat'].value_counts()
```

```python
df['cabin_cat'].value_counts().plot.bar()
```

```python
df['Ticket'].unique()
```

```python
# extract the last bit of ticket as number
df['ticket_num'] = df['Ticket'].apply(lambda s: s.split()[-1])
df['ticket_num'] = pd.to_numeric(df['ticket_num'],
                                   errors='coerce',
                                   downcast='integer')

# extract the first part of ticket as category
df['ticket_cat'] = df['Ticket'].apply(lambda s: s.split()[0])
df['ticket_cat'] = np.where(df['ticket_cat'].str.isdigit(), np.nan,
                              df['ticket_cat'])

df.head(20)
```
